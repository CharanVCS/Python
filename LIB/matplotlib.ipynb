```python
from matplotlib import pyplot as plt 
```

```python
x = [1, 2, 3 , 4, 5]
y = [0, 5, 0, 10, 0]
```

Ploting graph and labing

```python
plt.plot(x, y)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

subplot graph

```python
plt.subplot(2,2,1)
plt.plot(x, y)
plt.title('1st subplot')

# plt.figure()
plt.subplot(2,2,2)
plt.plot(y, x)
plt.title('2st subplot')
plt.show()
```

Bar graph

```python
plt.bar(x, [1,5,1,10,1])
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

sin and cosin waves

```python
import numpy as np

x = np.linspace(-np.pi, np.pi, 15)
sin = np.sin(x)
cos = np.cos(x)

ax = plt.axes([0.1, 0.1, 0.8, 0.8])

# 'bs:' mentions blue color, square, dotted line
ax1 = ax.plot(x, sin, 'bs:')
# 'ro-' mentions red color, circles, solid line
ax2 = ax.plot(x, cos, 'ro-')

ax.legend(labels = ('sin function', 'cosin function'), loc ='upper left')

ax.set_title('trigonometric functions')

plt.show()
```
